## Taylor series of $ln(x+1)$

$$ ln(x+1) = \sum_{n=0}^{\infty} \frac{(-1)^nx^n}{n} $$

What is value of the partial Taylor series of $ln(x+1)$ up to the  $n$-th term of $ln(x+1)$ about $0$ (Maclaurin  series) for $x=2$

```python
import math

n = 10
x = 2
e_to_x = 0
for i in range(1, n+1):
    e_to_x += ((((-1)**i)*(x**i))/i)
    print('ln ≈ ∑_{i=1}^'+str(i),'(-1)^ix^i)/i =',e_to_x, )  
print('ln(x+1) = ',math.log(x+1))
```

## Taylor series of $sin(x)$

$$ sin(x) = \sum_{n=0}^{\infty} \frac{(-1)^nx^{2n+1}}{(2n+1)!} $$

What is value of the partial Taylor series of $sin(x)$ up to the  $n$-th term of $sin(x)$ about $0$ (Maclaurin  series) for $x=2$

```python
import math

n = 10
x = 2
e_to_x = 0
for i in range(1, n+1):
    e_to_x += -((((-1)**i)*(x**(2*i+1)))/math.factorial(i*2+1))
    print('sin ≈ ∑_{i=1}^'+str(i),'(-1)^ix^(2i+1))/(2i+1)! =',e_to_x, )  
print('sin(x) = ',math.log(x+1))
```

What is the minimun $n$ for which the $n+1$-th term of the Taylor series for $sin(x)$ about $0$ (Maclaurin  series) for $x$, is in absolute value less than  $\epsilon$.

$$\min_{n \in \mathbb{N}}\left| \frac{(-1)^nx^{2n+1}}{(2n+1)!} \right| < \epsilon$$
